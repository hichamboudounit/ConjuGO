```python
def get_indicatif(lien):
    
    request_indicatif = requests.get(lien)
    html_indicaif =  BeautifulSoup(request_indicatif.content, 'html.parser')
    h2_indicatif = html_indicaif.find("h2", id = "ind")
    
    temps = {}

    for sibling in h2_indicatif.find_next_siblings():
        temps_key = ""
        conjugaison = []
        if sibling.name == 'div':
            conjug = ""
            for child in sibling.children:
                if child.name == 'br':
                    conjugaison.append(conjug)
                    conjug = ""
                elif child.name == 'div':
                    temps_key = child.get_text().strip()
                    conjug = ""
                else : 
                    conjug = conjug + child.get_text().strip()

            conjugaison.append(conjug)
            temps[temps_key] = conjugaison
        
        else:
            break

    return temps
    
```

```python
def get_subjonctif(lien):
    
    request_subjonctif = requests.get(lien)
    html_subjonctif =  BeautifulSoup(request_subjonctif.content, 'html.parser')
    h2_subjonctif = html_subjonctif.find("h2", id = "sub")
    
    temps = {}

    for sibling in h2_subjonctif.find_next_siblings():
        temps_key = ""
        conjugaison = []
        if sibling.name == 'div':
            conjug = ""
            for child in sibling.children:
                if child.name == 'br':
                    conjugaison.append(conjug)
                    conjug = ""
                elif child.name == 'div':
                    temps_key = child.get_text().strip()
                    conjug = ""
                elif child.name == 'b':
                    conjug = conjug + child.get_text().strip()
                else : 
                    conjug = conjug + child.get_text().strip()

            conjugaison.append(conjug)
            temps[temps_key] = conjugaison
        
        else:
            break

    return temps
    
```

```python
def get_imperatif(lien):
    
    request_imperatif = requests.get(lien)
    html_imperatif =  BeautifulSoup(request_imperatif.content, 'html.parser')
    h2_imperatif = html_imperatif.find("h2", id = "imp")
    
    temps = {}

    for sibling in h2_imperatif.find_next_siblings():
        temps_key = ""
        conjugaison = []
        if sibling.name == 'div':
            conjug = ""
            for child in sibling.children:
                if child.name == 'br':
                    conjugaison.append(conjug)
                    conjug = ""
                elif child.name == 'div':
                    temps_key = child.get_text().strip()
                    conjug = ""
                elif child.name == 'b':
                    conjug = conjug + child.get_text().strip()
                else : 
                    conjug = conjug + child.get_text().strip()

            conjugaison.append(conjug)
            temps[temps_key] = conjugaison
        
        else:
            break

    return temps
    
```

```python
def get_conditionnel(lien):
    
    request_conditionnel = requests.get(lien)
    html_conditionnel =  BeautifulSoup(request_conditionnel.content, 'html.parser')
    h2_conditionnel = html_conditionnel.find("h2", id = "cnd")
    
    temps = {}

    for sibling in h2_conditionnel.find_next_siblings():
        temps_key = ""
        conjugaison = []
        if sibling.name == 'div':
            conjug = ""
            for child in sibling.children:
                if child.name == 'br':
                    conjugaison.append(conjug)
                    conjug = ""
                elif child.name == 'div':
                    temps_key = child.get_text().strip()
                    conjug = ""
                elif child.name == 'b':
                    conjug = conjug + child.get_text().strip()
                else : 
                    conjug = conjug + child.get_text().strip()

            conjugaison.append(conjug)
            temps[temps_key] = conjugaison
        
        else:
            break

    return temps
    
```

```python
def main_extractor(verbe, lien):
    dict_verbe = {}
    dict_verbe['verbe'] = verbe
    dict_verbe['indicatif'] = get_indicatif(lien)
    dict_verbe['subjonctif'] = get_subjonctif(lien)
    dict_verbe['conditionnel'] = get_conditionnel(lien)
    dict_verbe['impératif'] = get_imperatif(lien)

    return dict_verbe
```

```python
data = []
for vrb, link in list(verbs_dict.items())[:10]:
    data.append(main_extractor(vrb, link))
```

```python
with open("Conjugaison.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Fichier JSON généré !")
```
