```python
def get_indicatif(lien):
    
    request_indicatif = requests.get(lien)
    html_indicaif =  BeautifulSoup(request_indicatif.content, 'html.parser')
    h2_indicatif = html_indicaif.find("h2", id = "ind")
    
    temps = {}
    mode = 'Indicatif'

    for sibling in h2_indicatif.find_next_siblings():
        temps_key = ""
        conjugaison = []
        if sibling.name == 'div':
            conjug = ""
            for child in sibling.children:
                if child.name == 'br':
                    conjugaison.append(conjug)
                    conjug = ""
                elif child.name == 'div':
                    temps_key = child.get_text().strip()
                    conjug = ""
                elif child.name == 'b':
                    conjug = conjug + child.get_text().strip()
                else : 
                    conjug = conjug + child.get_text().strip()

            temps[temps_key] = conjugaison
        
        else:
            break

    return {'indicatif' : temps}
    
```

```python
data = {}

request_abaisser = requests.get('https://leconjugueur.lefigaro.fr//conjugaison/verbe/abaisser.html', 'html.parser')
html_abaisser =  BeautifulSoup(request_abaisser.content, 'html.parser')
# Extraction de indicatif

h2_indicatif = html_abaisser.find("h2", id = "ind")
temps = {}
mode = 'Indicatif'

for sibling in h2_indicatif.find_next_siblings():
    temps_key = ""
    conjugaison = []
    if sibling.name == 'div':
        conjug = ""
        for child in sibling.children:
            if child.name == 'br':
                conjugaison.append(conjug)
                conjug = ""
            elif child.name == 'div':
                temps_key = child.get_text().strip()
                conjug = ""
            elif child.name == 'b':
                conjug = conjug + child.get_text().strip()
            else : 
                conjug = conjug + child.get_text().strip()

        temps[temps_key] = conjugaison
        data[mode] = temps
        
        
    else:
        break

```

```python
with open("Conjugaison.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Fichier JSON généré !")
```
